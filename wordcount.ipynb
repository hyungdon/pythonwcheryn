```python
import matplotlib
```

```python
lyrics = "i parked my car in harvard yard"
        # This is lyrics\n",
searchword = "car"
def wordmatch (lyrics, searchword):
    #\"find the searchword in lyrics\"\n",
    width = len(searchword)
    bindingscore = []
    for i in range(len(lyrics)-width):
        score = 0
        for j in range(width):
            if searchword[j] == lyrics [i+j]:
                score += 1
                bindingscore.append (score)
    return bindingscore
print (wordmatch (lyrics, searchword))
wordcount = wordmatch (lyrics, searchword)

from matplotlib import pyplot
pyplot.plot (wordcount)
pyplot.ylim (0, 3)
pyplot.show ()
```
